## Random Collisions

```python
def create_string_array(count, letter):
    array_values = []
    for i in range(count):
        array_values.append(letter)
    
    return array_values

def create_arrays(num_A, num_B, num_C):

    global array_a
    array_a = create_string_array(num_A, 'A')
    global array_b 
    array_b = create_string_array(num_B, 'B')
    global array_c 
    array_c = create_string_array(num_C, 'C')
```

```python
def collide_particles3(order, array_a, array_b, array_c):  
    # Define the possible collisions    
    collisions = {
        "ab": (array_a, array_b, array_c, 'C'),
        "ac": (array_a, array_c, array_b, 'B'),
        "bc": (array_b, array_c, array_a, 'A')
    }

    # Create a list of operations from the order string
    ops = [order[i:i+2] for i in range(0, len(order), 2)]

    # Store the number of A particles
    num_A = []
    num_B = []
    num_C = []

    # Simulate the collisions
    for op in ops:
        array1, array2, array3, letter = collisions[op]
        if array1 and array2:
            array1.pop(0)
            array2.pop(0)
            array3.extend([letter,letter])
            
            current_state_A = (array_a.copy())
            current_state_B = (array_b.copy())
            current_state_C = (array_c.copy())
            
        num_A.append(len(current_state_A))
        num_B.append(len(current_state_B))
        num_C.append(len(current_state_C))
    return num_A, num_B, num_C
```

This code generates random collisions

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def generate_random_collisions(num_collisions):
    # Define the possible collisions
    collisions = ["ab", "ac", "bc"]
    
    # Generate a list of random collisions
    random_collisions = [random.choice(collisions) for _ in range(num_collisions)]
    
    return "".join(random_collisions)
```

```python
def create_string_array(count, letter):
    array_values = []
    for i in range(count):
        array_values.append(letter)
    
    return array_values

def create_arrays2(num_A, num_B, num_C, num_D):

    global array_a
    array_a = create_string_array(num_A, 'A')
    global array_b 
    array_b = create_string_array(num_B, 'B')
    global array_c 
    array_c = create_string_array(num_C, 'C')
    global array_d 
    array_d = create_string_array(num_D, 'D')
```

```python
def collide_particles4(order, array_a, array_b, array_c, array_d):  
    # Define the possible collisions    
    collisions = {
        "ab": (array_a, array_b, array_c, array_d, 'C','D'),
        "ac": (array_a, array_c, array_b, array_d,'B','D'),
        "bc": (array_b, array_c, array_a, array_d,'A','D'),
        "ad": (array_a, array_d, array_b, array_c,'B','C'),
        "bd": (array_b, array_d, array_a, array_c, 'A','C'),
        "cd": (array_c, array_d, array_a, array_b,'A','B')
    }

    # Create a list of operations from the order string
    ops = [order[i:i+2] for i in range(0, len(order), 2)]

    # Store the number of A particles
    num_A = []
    num_B = []
    num_C = []
    num_D = []

    # Simulate the collisions
    for op in ops:
        array1, array2, array3, array4, letter1, letter2 = collisions[op]
        if array1 and array2:
            array1.pop(0)
            array2.pop(0)
            array3.extend([letter1])
            array4.extend([letter2])
            
            current_state_A = (array_a.copy())
            current_state_B = (array_b.copy())
            current_state_C = (array_c.copy())
            current_state_D = (array_d.copy())
            
        num_A.append(len(current_state_A))
        num_B.append(len(current_state_B))
        num_C.append(len(current_state_C))
        num_D.append(len(current_state_D))
    return num_A, num_B, num_C, num_D
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def generate_random_collisions2(num_collisions):
    # Define the possible collisions
    collisions = ["ab", "ac", "bc", "ad", "bd","cd"]
    
    # Generate a list of random collisions
    random_collisions = [random.choice(collisions) for _ in range(num_collisions)]
    
    return "".join(random_collisions)
```
