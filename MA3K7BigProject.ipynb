```python
def create_string_array(count, letter):
    array_values = []
    for i in range(count):
        array_values.append(letter)
    
    return array_values

def create_arrays(num_A, num_B, num_C):

    global array_a
    array_a = create_string_array(num_A, 'A')
    global array_b 
    array_b = create_string_array(num_B, 'B')
    global array_c 
    array_c = create_string_array(num_C, 'C')
```

This code creates all possible orders of X amount of collisions, and prints either:
- We return back to the starting case.
- We end up with only one species.
- We return back somewhere on route.
- If none of these are the case, no result.

```python
def collide_particles1(order, array_a, array_b, array_c):  
    # Define the possible collisions    
    collisions = {
        "ab": (array_a, array_b, array_c, 'C'),
        "ac": (array_a, array_c, array_b, 'B'),
        "bc": (array_b, array_c, array_a, 'A')
    }

    # Create a list of operations from the order string
    ops = [order[i:i+2] for i in range(0, len(order), 2)]

    # Store the initial state of the arrays
    initial_state = (array_a.copy(), array_b.copy(), array_c.copy())
    previous_states = [initial_state]

    # Simulate the collisions
    for op in ops:
        array1, array2, array3, letter = collisions[op]
        if array1 and array2:
            array1.pop(0)
            array2.pop(0)
            array3.extend([letter,letter])

            # Store the current state of the arrays
            current_state = (array_a.copy(), array_b.copy(), array_c.copy())
            # Check if there are at least two species present
            species_count = sum(bool(array) for array in current_state)
            if species_count >= 2:
                previous_states.append(current_state)
        else:
            print("Unable to perform collision")
    if current_state == initial_state:
        print("Returned to starting case")
    elif current_state in previous_states[:-1]:
        print("Returned somewhere in the route")
    elif len(set(array_a + array_b + array_c)) == 1:
        print("Only one species remaining")
    else:
        print("No result")
    print()
```

```python
from itertools import product
# Define the possible orders of collisions
orders = ['ab', 'ac', 'bc']
# Generate all possible orders of 3 collisions
order = [''.join(combination) for combination in product(orders, repeat=3)]
```

```python
for order in order:
    print(f"For order {order}:")
    create_arrays(2, 2, 2)
    collide_particles1(order, array_a, array_b, array_c)
```

This code creates all possible orders of X amount of collisions, and prints either:
- All collisions were performed and we got a result.
- All collisions were performed but we didn't get a result.
- Not all collisions were performed but we got a result.

```python
def collide_particles2(order, array_a, array_b, array_c):  
    # Define the possible collisions    
    collisions = {
        "ab": (array_a, array_b, array_c, 'C'),
        "ac": (array_a, array_c, array_b, 'B'),
        "bc": (array_b, array_c, array_a, 'A')
    }

    # Create a list of operations from the order string
    ops = [order[i:i+2] for i in range(0, len(order), 2)]

    # Store the initial state of the arrays
    initial_state = (array_a.copy(), array_b.copy(), array_c.copy())
    previous_states = [initial_state]

    # Variables to track the status of the collisions
    all_collisions_done = True
    result_obtained = False

    # Simulate the collisions
    for op in ops:
        array1, array2, array3, letter = collisions[op]
        if array1 and array2:
            array1.pop(0)
            array2.pop(0)
            array3.extend([letter,letter])

            # Store the current state of the arrays
            current_state = (array_a.copy(), array_b.copy(), array_c.copy())
            # Check if there are at least two species present
            species_count = sum(bool(array) for array in current_state)
            if species_count >= 2:
                previous_states.append(current_state)
                result_obtained = True
        else:
            all_collisions_done = False
            #print("Unable to perform collision")

    # Print the desired outputs based on the status of the collisions and the result
    if all_collisions_done and result_obtained:
        print(f"For order {order}: All collisions were performed and we got a result.")
    elif all_collisions_done and not result_obtained:
        print(f"For order {order}: All collisions were performed but we didn't get a result.")
    elif not all_collisions_done and result_obtained:
        print(f"For order {order}: Not all collisions were performed but we got a result.")
```

```python
# Generate all possible orders of 3 collisions
order = [''.join(combination) for combination in product(orders, repeat=3)]
for order in order:
    create_arrays(2, 2, 2)
    collide_particles2(order, array_a, array_b, array_c)
```
